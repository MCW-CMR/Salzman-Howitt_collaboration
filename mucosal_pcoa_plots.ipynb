# Make mucosal PCoA plots

```python
#import libraries
import pandas as pd
import numpy as np 
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
#import seaborn as sns
```

```python
#create figure
my_dpi = 300
plt.figure(figsize=(12, 6), dpi=my_dpi)
```

```python
#load data and make df
df = pd.read_csv("exported_bray_pcoa_results/bray_pcoa_results_mucosal.csv", index_col=0)
df
```

```python
#initialize plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
marker = {0:'*',1:'.'}
ax.scatter(df['PC1'], df['PC3'], df['PC2'], c=df['color'], depthshade=False, s=60, edgecolor='black')
# label the axes
ax.grid(False)
ax.view_init(15, 65)
ax.set_xlabel("PC1 (61.56%)")
ax.set_zlabel("PC2 (19.83%)")
ax.set_ylabel("PC3 (7.06%)")
#legend
blue_dot = plt.Line2D([], [], color="blue", ls="", marker='.', markersize=15, label = "PBS")
yellow_dot = plt.Line2D([], [], color="yellow", ls="", marker='.', markersize=15, label = "IL25")
#green_dot = mpatches.Patch(color="darkgreen", label = "pre-op")
#yellow_dot = mpatches.Patch(color="yellow", label = "post 3m")

plt.legend(handles=[blue_dot, yellow_dot], loc='upper left', bbox_to_anchor=(0.7, 1))#
#plt.show()
plt.savefig('bray_pcoa_mucosal.png', format='png', dpi = 500, bbox_inches="tight")
```
